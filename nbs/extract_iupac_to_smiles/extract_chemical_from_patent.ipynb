# Extract IUPAC from patent

## 31678

```python
import pandas as pd
```

```python
df = pd.read_csv('data/WO2022031678.csv',header=None)
```

```python
df
```

```python
# Initialize lists for example numbers and chemical names
example_numbers = []
chemical_names = []

# Loop through the DataFrame
for i, row in df.iterrows():
    if "EXAMPLE" in row[0]:
        example_number = row[0].split()[1]
        example_numbers.append(example_number)
        
        # Assuming the chemical name is in the next row
        chemical_name = df.loc[i + 2, 0]
        chemical_names.append(chemical_name)

# Create a new DataFrame with the extracted data
extracted_df = pd.DataFrame(
    {"Example": example_numbers, "Chemical Name": chemical_names}
)
```

```python
extracted_df
```

```python
extracted_df.to_csv('data/iupac_WO2022031678.csv')
```

## Aza

```python
import pandas as pd
```

```python
df = pd.read_csv('data/aza_patent.csv',header=None)
```

```python
df
```

```python
# Initialize lists for example numbers and chemical names
example_numbers = []
chemical_names = []

# Loop through the DataFrame
for i, row in df.iterrows():
    try:
        if row[0].lower().startswith('example'):
            example_number = row[0].split()[1]
            example_numbers.append(example_number)
            
            print(row[0])

            # Assuming the chemical name is in the next row
            chemical_name = df.loc[i + 1, 0]
            chemical_names.append(chemical_name)
    except Exception as e:
            print(f"Error at index {i}: {e}")
            continue

# Create a new DataFrame with the extracted data
extracted_df = pd.DataFrame(
    {"Example": example_numbers, "IUPAC": chemical_names}
)
```

```python
extracted_df
```

```python
extracted_df.to_csv('data/iupac_aza.csv')
```
